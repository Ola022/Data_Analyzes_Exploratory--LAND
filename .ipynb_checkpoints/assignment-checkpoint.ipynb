#### NOTE: ANYONE AGED `OVER 18 CAN BA A HEAD`
+ ` - From the 'About' given, it was stated that `Head` will only be applicable to the occupant aged over 18`
+ ` - And only `Head` has the responsibility to fill in the census details;.`
+ ` - Anyone aged “over 18” – that is greater than 18 (>18). `

```python
# filtered the occupant of 17 years or lower that is Head of the House (outliners )

filtered_under_age_head_df = df[(df['Age'] <= 18) & (df['Relationship_to_Head'] == 'Head')] 
filtered_under_age_head_df.head(3)
```

```python
# - Find the lenght of filtered under-age Head DataFrame
len(filtered_under_age_head_df)
```

#### CHECK DATA FILLED BY UNDER-AGE HEAD

```python
# Check data filled is by uder-age head by comparing there 'House_Number' and 'Street' column of the data filled
def check(row):    
    for x in filtered_under_age_head_df[['House_Number','Street']].values:
        if (row['House_Number'], row['Street']) == tuple(x):
            return True
    return False
    
underage_head_filleddata_df = df[df.apply(check, axis=1)]
underage_head_filleddata_df.head(10)
# df[(df['House_Number'].isin(house_numbers_to_include) & df['Street'].isin(street_column_to_include))]

```

```python
# - Lenght of filtered under-age Head of House and data filled 
len(underage_head_filleddata_df)
```

#### REMOVE UNDERAGE HEAD OF THE HOUSE and OTHER DATA FILLED

```python
# Identify rows in 'df' where the values exist in 'original_df'.
mask = df.isin(underage_head_filleddata_df)
mask.sum()

# Use the 'any' method to check if any value in a row exists in 'original_df'.
mask = mask.any(axis=1)

# Drop rows from 'df' where the values exist in 'original_df'.
df_2 = df.drop(df[mask].index)
# df_2 = df 
df_2.tail(8)
```
